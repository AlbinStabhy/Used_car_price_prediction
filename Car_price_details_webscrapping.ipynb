```python
import selenium
import pandas as pd
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome(r"C://chromedriver.exe")
```

```python
driver.get('https://www.olx.in/items/q-cars?isSearchCall=true')  # Cars in India
```

```python
driver.maximize_window()
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
links = []
```

```python
for link in links_tags:
    lnk = link.get_attribute(name = 'href')
    links.append(lnk)
links
```

```python
len(links)
```

```python
links  = list(dict.fromkeys(links)) # Remove the duplicates
```

```python
driver.get('https://www.olx.in/rajasthan_g2001171/q-cars')
```

```python
scrollnumber = 500
try:
    for i in range(1,scrollnumber):
        driver.execute_script("window.scrollTo(0,50000)")
        driver.find_element_by_xpath("//li[@class ='_20FqS']/div/button").click()
        time.sleep(3)
        links_tags = driver.find_elements_by_xpath("//div[@class = '_3etsg']/div/div[2]/ul/li/a")
except NoSuchElementException:
    pass
```

```python
for link in links_tags:
    lnk = link.get_attribute(name = 'href')
    links.append(lnk)
links
```

```python
len(links)
```

```python
links  = list(dict.fromkeys(links))
```

```python
len(links)
```

```python
brand = []
model = []
kilometers = []
variant = []
year = []
fuel = []
no_of_owners = []
price = []
```

```python
for link in links:
    driver.get(link)
    time.sleep(2)
    
    try:
        brand.append(driver.find_element_by_xpath('//span[@data-aut-id = "value_make"]').text)
    except NoSuchElementException:
        brand.append("--")
        
    try:
        model.append(driver.find_element_by_xpath('//span[@data-aut-id = "value_model"]').text)
    except NoSuchElementException:
        model.append("--")
        
    try:
        variant.append(driver.find_element_by_xpath("//span[@data-aut-id = 'value_variant']").text)
    except NoSuchElementException:
        variant.append("--")
        
    try:
        year.append(driver.find_element_by_xpath("//span[@data-aut-id= 'value_year']").text)
    except NoSuchElementException:
        year.append("--")
        
    try:
        fuel.append(driver.find_element_by_xpath("//span[@data-aut-id ='value_petrol']").text)
    except NoSuchElementException:
        fuel.append("--")
        
    try:
        kilometers.append(driver.find_element_by_xpath("//span[@data-aut-id = 'value_mileage']").text)
    except NoSuchElementException:
        kilometers.append("--")
        
    try:
        no_of_owners.append(driver.find_element_by_xpath("//span[@data-aut-id ='value_first_owner']").text)
    except NoSuchElementException:
        no_of_owners.append("--")
        
    try:
        price.append(driver.find_element_by_xpath('//section[@class = "_2wMiF"]/span[1]').text)
    except NoSuchElementException:
        price.append("--")
```

```python
olx = pd.DataFrame({})
olx['Brand'] = brand
olx['Model'] = model
olx['Year'] = year
olx['Kilometers'] = kilometers
olx['No_of_owners'] = no_of_owners
olx['Fuel'] = fuel
olx['Variant'] = variant
olx['Price'] = price
```

```python
driver.get('https://www.cardekho.com/used-cars+in+chennai')
```

```python
links1 = []
```

```python
variant1 = []
```

```python
no_of_owners1 = []
year1 = []
kilometers1 = []
fuel1 = []
price1 = []
```

```python
try:
    scrollnumber = 500
    for i in range(1,scrollnumber):
        driver.execute_script("window.scrollTo(0,40000)")
        time.sleep(3)
        links1_tags = driver.find_elements_by_xpath('//div[@class = "gsc_col-xs-7 carsName"]/a')
        variant1_tags = driver.find_elements_by_xpath('//div[@class = "gsc_col-xs-7 carsName"]/div[1]')
    
except NoSuchElementException:
    pass
```

```python
for link in links1_tags:
    links1.append(link.get_attribute(name='href'))
    
links1
```

```python
len(links1)
```

```python
for var in variant1_tags:
    variant1.append(var.text)
    
variant1
```

```python
len(variant1)
```

```python
brand_n_model = []
```

```python
for link in links1:
    driver.get(link)
    time.sleep(3)
    
    try:
        brand_n_model.append(driver.find_element_by_xpath('//div[@class = "vdp-head"]/h1[2]').text)
    except NoSuchElementException:
        brand_n_model.append("--")
        
        
    try:
        no_of_owners1.append(driver.find_element_by_xpath('//div[@class = "GenDetailBox"]/ul/li[4]').text)
    except NoSuchElementException:
        no_of_owners1.append('--')
        
    try:
        year1.append(driver.find_element_by_xpath('//div[@class = "GenDetailBox"]/ul/li[1]').text)
    except NoSuchElementException:
        year1.append('--')
        
    try:
        kilometers1.append(driver.find_element_by_xpath('//div[@class = "GenDetailBox"]/ul/li[3]').text)
    except NoSuchElementException:
        kilometers1.append('--')
        
    try:
        fuel1.append(driver.find_element_by_xpath('//div[@class = "GenDetailBox"]/ul/li[5]').text)
    except NoSuchElementException:
        fuel1.append('--')
        
    try:
        price1.append(driver.find_element_by_xpath('//span[@class = "pricewrapper posR"]/span').text)
    except NoSuchElementException:
        price1.append('--')
```

```python
cardheko = pd.DataFrame({})
cardheko['Brand_n_Model'] = brand_n_model
cardheko['Variant'] = variant1
cardheko['Year'] = year1
cardheko['Kilometers'] = kilometers1
cardheko['No_of_owners'] = no_of_owners1
cardheko['Fuel'] = fuel1
cardheko['Price'] = price1
```

```python
print(len(brand_n_model))
print(len(variant1))
print(len(year1))
print(len(kilometers1))
print(len(no_of_owners1))
print(len(fuel1))
print(len(price1))
```
